## Importing spacy library and VADER module

```python
import pandas as pd                                    
import spacy

from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

## Importing and reading CSV file 

```python
df=pd.read_csv("amazonreviews.csv")
df.head()
```

## Getting polarity scores for the reviews
* The VADER module returns probablities of the text being positive, negative or neutral and a compound score
* The compound score ranges from -1(most negative) to 1 (most positive) and 0 being neutral
* The polarity of the reviews are classified on the basis of compound scores

```python
scores = []
for reviews in df['reviews.text']:
    scores.append(sid.polarity_scores(str(reviews)).get('compound'))

df['Polarity'] = scores
```

## Defining a function for assigning positive,negative and neutral tags

```python
def polarity_val(Polarity):
    if Polarity>0.1:
        return 'Positive'
    elif Polarity<-0.1:
        return 'Negative'
    else:
        return 'Neutral'

df['Polarity']=df['Polarity'].apply(polarity_val)
```
